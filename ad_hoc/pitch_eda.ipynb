```python
from pyspark.sql import functions as F

pitch_df = spark.read.table("mlb.03_gold.pitcher_yearly")
```

## Dominance in the Zone (Whiff vs Zone Rate)

```python
filtered_df = pitch_df.filter(F.col("pitch_count") > 50)
filtered_df.display()
```

## Dominance pitchers (2025)

```python
fastball_dominance_df = pitch_df.filter(
    (F.col("pitch_name") == "4-Seam Fastball") & 
    (F.col("pitch_count") > 1000) & 
    (F.col("Season") == "2025")
).withColumn(
    "raw_stuff", (F.col("avg_velocity") * 0.4) + (F.col("total_break") * 0.6)
).withColumn(
    "dominance_score",
    F.round(
        F.col("raw_stuff") * F.col("whiff_rate") * (1.5 - F.col("slugging_per_pitch")), 
        2
    )
)

top_dominance_pitchers = fastball_dominance_df.orderBy(F.desc("dominance_score")).limit(20)
```
